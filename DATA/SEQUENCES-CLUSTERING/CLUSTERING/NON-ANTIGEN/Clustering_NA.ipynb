### This bash script runs 'easy-cluster' module in mmseqs. It clusters the sequences with minimum sequence identity of 30% and coverage of 80%. 

```python
%%bash 

mkdir tmp
mmseqs easy-cluster ../../FILTERING/filtered_non_antigens.fasta NA_clu_result tmp --min-seq-id 0.3 -c 0.8 --cov-mode 0 --cluster-reassign -v 1
```
